# Contexto y objetivo

*The Mexican Health and Aging Study (MHAS) is a national longitudinal study of adults 50 years and older in Mexico*.

**Why was the cohort set up?**

By 2050, adults aged 60 years and older **will be the largest group** in the population age pyramid of most countries.

The MHAS (Mexican Health and Aging Study) is partly sponsored by the **National Institutes of Health/National Institute on Aging** (grant number NIH R01AG018016) in the United States and the **Instituto Nacional de Estadística y Geografía** (INEGI) in Mexico.

The Mexican Health and Aging Study (MHAS) was designed to prospectively **evaluate the impact of disease on the health, function and mortality** of adults over the age of 50 in both **urban and rural areas of Mexico**.

The **overall goal** of the study is **to examine the ageing process** and **its disease and disability burden** in a large representative panel of older Mexicans, using a *wide socioeconomic* perspective.

#  ¿De dónde proviene?

*The Mexican Health and Aging Study (MHAS) is **a national longitudinal study** of adults 50 years and older in Mexico*.

The survey has national and urban/rural representation.

The baseline survey, in 2001, included a nationally representative sample of Mexicans aged 50 and over and their spouse/partners regardless of their age.

The sample was distributed in **all 32 states** of the country in **urban** and **rural** areas. Households in the **six states** which account for **40%** of all **migrants** to the U.S. **were over-sampled**.



* A follow-up interview with surviving respondents was conducted in 2003.
* A third follow-up survey that was conducted in 2012.
* In 2012 a new sample of persons was added.
* A fourth round of the longitudinal study was completed in 2015.
* A fifth follow-up survey that was conducted in 2018.
* In 2018 a new sample of persons was added to the sample.
* A fifth round of the longitudinal study was completed in 2021.
* A seventh follow-up survey that was conducted in 2024.
* In 2024 a new sample of persons was added.
* An eighth round of the longitudinal study is planned for 2027.


Since 2003, the following has been carried out in each wave of study:

* A follow-up interview with all surviving respondents,
* A next-of-kin interview about deceased respondents;
* A complete baseline interview for new spouse/partners,
* Direct or proxy interviews for new samples,
* A follow-up proxy interview for respondents unable to complete their own interview because of illness or temporary absence.

# ¿Es real o simulado?

* A direct interview was sought with each individual and proxy interviews were obtained when poor health or temporary absence precluded a direct interview.

* All interviews were conducted in person with paper-pencil by trained full-time interviewers.

* From 2012 and on Interviews were conducted person-to-person using CAPI (Computer-Assisted Personal Interviewing).

# ¿Qué fenómeno describe el dataset?
The overall goal of the study is to examine the **ageing process** and **its disease and disability burden** in a large representative panel **of older Mexicans**, using a wide **socioeconomic** perspective.

*Survey Contents*

**The MHAS content includes**:

* Health in multiple domains (self-report of global health, chronic conditions, symptom reports, functionality, depression, cognition).
* Socioeconomic conditions (current and in childhood), work history, health insurance, health expenditures.
* Family background (family structure, transfer behaviors, care arrangements, health and migration histories of respondents, parents and children), children (regardless of place of residence) and household residents.
* Income, assets, pension history, current housing, and quality of built environment.
* Major events over the last 10 years (only in 2012): change of residence, major health event, natural disaster, or crime event.
* Time use and psychosocial aspects (locus of control, life satisfaction, loneliness, and conscientiousness).
* For deceased study participants, last year of life: major events, use of health services, care receive, and functionality.
* Sub-sample for biomarkers, anthropometrics, and performance measures (only in 2012).

# Unidad de análisis

The unit of analysis is the *sampled* **individual respondent** (person-level data).

The MHAS public-use files are structured at the individual level. So, Default Level: *Individual*.

In MHAS Section A-Demographics, these two concepts align: the unit of analysis (the person) matches the level of aggregation (individual rows).

What is measured?: Characteristics, experiences, and behaviors of a single older adult (age 50+).

Even household-level variables (like home value) are attributed to each individual in that household. So each person gets their household's wealth value.

Proxy responses: When a proxy answers, the unit remains the target respondent, not the proxy.

What happens with proxy responses?

* Proxy responses are handled in two distinct ways:
    * A. Proxy for Living Respondents (Section PC). When the sampled person cannot answer due to health/cognitive issues:
What's measured: The proxy answers for the respondent. The unit of analysis remains the sampled person, but the source is the proxy.

    * Next of Kin (NK) for Deceased Respondents. When the sampled person died since last wave: What's measured: The NK informant provides information about the deceased. The unit of analysis is still the original sampled person, but now with end-of-life data.

Critical distinction:

Proxy (PC) = Living person, someone else answers, Still the same unit, different data source.

NK = Deceased person, someone else answers about the past,Still the same unit, retrospective data.

```python
def codebook_style_summary(df):
    """Generate summary in style of Stata codebook"""

    for col in df.columns:
        print("\n" + "="*60)
        print(f"{col} ({df[col].dtype})")
        print("="*60)

        # Basic info
        print(f"Observations: {len(df)}")
        print(f"Non-missing: {df[col].count()}")
        print(f"Missing: {df[col].isna().sum()} ({df[col].isna().sum()/len(df)*100:.1f}%)")

        if df[col].dtype in ['object', 'category']:
            # Categorical summary
            print(f"Unique values: {df[col].nunique()}")
            print("\nValue counts:")
            value_counts = df[col].value_counts(dropna=False).head(10)
            for val, count in value_counts.items():
                pct = count/len(df)*100
                print(f"  {val}: {count} ({pct:.1f}%)")
        else:
            # Numerical summary
            print(f"Unique values: {df[col].nunique()}")
            print("\nValue counts:")
            value_counts = df[col].value_counts(dropna=False).head(10)
            for val, count in value_counts.items():
                pct = count/len(df)*100
                print(f"  {val}: {count} ({pct:.1f}%)")
            print(f"\nStats:")
            print(f"  Min: {df[col].min()}")
            print(f"  Mean: {df[col].mean():.2f}")
            print(f"  Max: {df[col].max()}")
            print(f"  Std: {df[col].std():.2f}")

            # Show distribution quintiles
            print("\nDistribution:")
            percentiles = [0, 25, 50, 75, 100]
            for p in percentiles:
                val = df[col].quantile(p/100)
                print(f"  {p}%: {val}")

# Use it
codebook_style_summary(df)
```
